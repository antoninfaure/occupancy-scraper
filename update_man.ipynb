```python
# repeat for each week of the semester (except the first one)
def create_schedule_semester(schedules, semester):
    semester_start_date = semester["start_date"]
    semester_end_date = semester["end_date"]
    semester_start_date += timedelta(days=7)

    skip_dates = semester["skip_dates"]

    map_day_to_int = {
        "lundi": 0,
        "mardi": 1,
        "mercredi": 2,
        "jeudi": 3,
        "vendredi": 4,
        "samedi": 5,
        "dimanche": 6,
    }

    semester_schedules = []
    current_date = semester_start_date
    while current_date < semester_end_date:
        if current_date in skip_dates:
            current_date += timedelta(days=1)
            continue

        for schedule in schedules:
            if current_date.weekday() != map_day_to_int[schedule["day"]]:
                continue

            # Parse start hour
            start_hour = datetime.strptime(str(schedule["start_hour"]), "%H").time()
            start_datetime = datetime.combine(current_date, start_hour)

            # Calculate end datetime based on duration
            end_hour = datetime.strptime(str(schedule["end_hour"]), "%H").time()
            end_datetime = datetime.combine(current_date, end_hour)

            semester_schedules.append(
                {
                    "course_id": schedule["course_id"],
                    "label": schedule["label"],
                    "room_ids": schedule["room_ids"],
                    "start_datetime": start_datetime,
                    "end_datetime": end_datetime,
                }
            )

        current_date += timedelta(days=1)
    return semester_schedules
```

```python
def create_schedules(db, schedules):
    # insert schedules (without room_id)
    new_schedules = []
    for schedule in schedules:
        new_schedules.append(
            {
                "course_id": schedule["course_id"],
                "label": schedule["label"],
                "start_datetime": schedule["start_datetime"],
                "end_datetime": schedule["end_datetime"],
                "available": True,
            }
        )

    try:
        db.course_schedules.insert_many(new_schedules)
    except Exception as e:
        print(e)


def create_bookings(db, schedules):
    # find the schedules in the db
    db_schedules = list(
        db.course_schedules.find(
            {"course_id": {"$in": [schedule["course_id"] for schedule in schedules]}}
        )
    )

    # create bookings (room_id paired with schedule_id)
    new_bookings = []
    for schedule in schedules:
        # find the schedule with the same label, start_datetime and end_datetime
        db_schedule = next(
            (
                db_schedule
                for db_schedule in db_schedules
                if db_schedule["label"] == schedule["label"]
                and db_schedule["start_datetime"] == schedule["start_datetime"]
                and db_schedule["end_datetime"] == schedule["end_datetime"]
            ),
            None,
        )
        if db_schedule is not None:
            for room_id in schedule["room_ids"]:
                new_bookings.append(
                    {
                        "schedule_id": db_schedule["_id"],
                        "room_id": room_id,
                        "available": True,
                    }
                )

    try:
        db.course_bookings.insert_many(new_bookings)
    except Exception as e:
        print(e)
```

```python
# delete all bookings and schedules for a given course
def delete_bookings_and_schedules(db, course_id):
    db.course_bookings.delete_many(
        {
            "schedule_id": {
                "$in": [
                    schedule["_id"]
                    for schedule in db.course_schedules.find({"course_id": course_id})
                ]
            }
        }
    )
    db.course_schedules.delete_many({"course_id": course_id})
```
